```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 45, 300)
y1 = np.sin(2 * x) * np.cos(x)
y2 = 0.5 * np.sin(2 * x) * np.cos(x +  np.pi / 2)
y3 = 0.75 * np.sin(2 * x) * np.cos(x + 3 * np.pi / 2)

plt.plot(x, y1, color='red', linewidth=1)
plt.plot(x, y2, color='green',linewidth=1.5)
plt.plot(x, y3, color='blue',linewidth=1)
plt.xlabel('x, rad')
plt.ylabel('y')
plt.title('plot y1=y1(x), y2=y2(x), y3=y3(x)')
plt.grid()
plt.legend('1,2,3', loc = 'lower left')
plt.show()
```

```python
from scipy.stats import norm
```

```python
fig = plt.figure(figsize=(20, 9))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

pos = 0
scale = 100

size1 = 1000
values1 = np.random.normal(pos, scale, size1)
ax1.hist(values1, 100, density=True, color='green')
x1 = np.linspace(pos - 3*scale, pos + 3*scale, 100)
ax1.plot(x1, norm.pdf(x1, pos, scale), 'r-', lw=3)
ax1.set_title('normal distribution_1')


size2 = 10000
values2 = np.random.normal(pos, scale, size2)
ax2.hist(values2, 100, density=True)
x2 = np.linspace(pos - 3*scale, pos + 3*scale, 100)
ax2.plot(x2, norm.pdf(x2, pos, scale), 'r-', lw=3)
ax2.set_title('normal distribution_2')


size3 = 100000
values3 = np.random.normal(pos, scale, size3)
ax3.hist(values3, 100, density=True, color='orange')
x3 = np.linspace(pos - 3*scale, pos + 3*scale, 100)
ax3.plot(x3, norm.pdf(x3, pos, scale), 'r-', lw=3)
ax3.set_title('normal distribution_3')


size4 = 1000000
values4 = np.random.normal(pos, scale, size4)
ax4.hist(values4, 100, density=True, color='skyblue')
x4 = np.linspace(pos - 3*scale, pos + 3*scale, 100)
ax4.plot(x4, norm.pdf(x4, pos, scale), 'r-', lw=3)
ax4.set_title('normal distribution_4')


plt.tight_layout()
plt.show()
```
